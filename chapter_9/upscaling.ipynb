```python
import torch
import torch.nn as nn
```

```python
# initialize a network m, with the nn.ConvTranspose2d method
m = nn.ConvTranspose2d(1, 1, kernel_size=(2,2), stride=2, padding=0)
```

```python
# initialize the input array and pass it thru the network
input = torch.ones(1, 1, 3, 3)
output = m(input)
output.shape
```
